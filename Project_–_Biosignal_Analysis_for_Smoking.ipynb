# **Logistic Regression**

```python
x = df[['gender', "triglyceride", "height(cm)","Gtp","hemoglobin","age","weight(kg)","waist(cm)","HDL","fasting blood sugar","relaxation","ALT","serum creatinine","LDL","systolic"]]
y = df["smoking"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
accuracy_score(y_test, y_pred)
```

```python
## Clsssification Report for the Logistic Regression
lr_cls_report = classification_report(y_test, y_pred)
print(lr_cls_report)
```

# **Decision Tree**

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)
```

```python
y_pred1 = dt.predict(x_test)
```

```python
print(f"Accuracy Score for Decision Tree: {accuracy_score(y_test, y_pred1)}")
```

```python
## Classification Report of DT
dt_cls_report = classification_report(y_test, y_pred1)
print(dt_cls_report)
```

**Bagging Algorithm – Bagging Classifier**



Bootstrap Aggregation or bagging involves taking multiple samples from the training dataset
(with replacement) and training a model for each sample.

```python
from sklearn.ensemble import BaggingClassifier
```

```python
Bagging_clf = BaggingClassifier(estimator=DecisionTreeClassifier(), n_estimators=1000, random_state=42)
Bagging_clf.fit(x_train, y_train).score(x_test, y_test)
y_pred2= Bagging_clf.predict(x_test)
print(classification_report(y_test,y_pred2))
```

**Bagging Algorithm – Extra Trees**

```python
from sklearn.ensemble import ExtraTreesClassifier
et = ExtraTreesClassifier(n_estimators=1000, random_state=42)
et.fit(x_train, y_train).score(x_test, y_test)
y_pred3 = et.predict(x_test)
print(classification_report(y_test,y_pred3))
```

# **Bagging Algorithm – Random Forest**

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=1000, random_state=42)
rf.fit(x_train, y_train)
y_pred4 = rf.predict(x_test)
print(classification_report(y_test,y_pred4))
```
